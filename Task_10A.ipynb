### TASK-10A: Please model the AD dataset using the SVM, Decision Tree, and Random Forest algorithms. Please apply train test split (%75-%25) Then grid search (hyperparameter tuning) Then classification report and confusion matrix (on the test dataset) Then compare the findings from the three algorithms.

```python
import numpy as np
import pandas as pd

data = pd.read_excel('Temporary_data3_Left_Right_Copy.xlsx')
data.head()
```

```python
from sklearn.model_selection import train_test_split
Cdrglob = data["CDRGLOB"]
data.drop(columns = ["CDRGLOB"], axis = 1, inplace = True)
X_train, X_test, y_train, y_test = train_test_split(data, Cdrglob, test_size=0.25, random_state=42)
```

**SVC iÃ§in**

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV

model = SVC()

param_grid = {'kernel': ['rbf','poly'],
              'shrinking':[True,False]}

grid = GridSearchCV(model, param_grid, refit=True, verbose=3)
grid.fit(X_train, y_train)
grid.best_params_
```

```python
grid_predictions = grid.predict(X_test)
print(confusion_matrix(y_test, grid_predictions))
print(classification_report(y_test, grid_predictions))
```

```python
from sklearn.tree import DecisionTreeClassifier

param_grid = {'max_depth': [1,5,10],
              'min_samples_split': [2,5,10],
              'min_samples_leaf': [1,5,10]}
grid = GridSearchCV(DecisionTreeClassifier(), param_grid, refit=True, verbose=3)
grid.fit(X_train, y_train)
grid.best_params_
```

```python
grid_predictions = grid.predict(X_test)
print(confusion_matrix(y_test, grid_predictions))
print(classification_report(y_test, grid_predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier

param_grid = {'n_estimators': [10, 100, 1000],
                'max_depth': [1,5,10],
                'min_samples_leaf': [1,5,10]}
grid = GridSearchCV(RandomForestClassifier(), param_grid, refit=True, verbose=3)
grid.fit(X_train, y_train)
grid.best_params_
```

```python
grid_predictions = grid.predict(X_test)
print(confusion_matrix(y_test, grid_predictions))
print(classification_report(y_test, grid_predictions))
```
